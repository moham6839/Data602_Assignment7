## Modify Multiple Column Names

```python
df.rename({"Pk": "Pick", "Tm": "Team", "G": "Games Played", "Yrs": "Years", "WS": "WinShares", "MPG": "Minutes Per Game", "BPM": "Plus/Minus"}, axis=1, inplace=True)
df
```

## Fixing Missing Data

```python
print(df.isnull().sum())
```

```python
fillna_new_df = df
fillna_new_df.College.fillna("No College", inplace=True)
fillna_new_df.Years.fillna(0, inplace=True)
fillna_new_df
```

## Creating New Columns

```python
fillna_new_df["Round"] = np.where(fillna_new_df["Pick"] <= 30, "1", "2")
fillna_new_df.head()
```

## Drop Columns

```python
fillna_new_df.drop(["DraftYr", "Rk", "playerurl", "VORP"], axis=1, inplace=True)
fillna_new_df
```

## Drop Row

```python
fillna_new_df.dropna(axis=0, inplace=True)
fillna_new_df
```

```python
print(fillna_new_df.isnull().sum())
```

## Sort Data Based on Multiple Variables

```python
sort_df = fillna_new_df.sort_values(["WinShares", "Minutes Per Game"], ascending=[False, True]).head(20)
sort_df
```

```python
sort_df["College"].value_counts()
```

## Filter Data Based on Some Condition

```python
# Creating subset of data from 2005 to 2021; resetting index

new_df = fillna_new_df.loc[df["DraftYear"] >= 2005]
new_df.reset_index(inplace=True, drop=True)
```

## Convert all string values to upper or lower case in one column

```python
new_df["Team"] = new_df["Team"].map(str.lower)
new_df
```

## Check whether numeric values are present in a given column of your dataframe.

```python
sum(new_df.Team.str.isnumeric())
```

## Group your dataset by one column, and get the mean, min, and max values by group.

```python
results = new_df.groupby("Round").agg({"WinShares": ["mean", "min", "max"]})
print("Mean, min, and max values of Points Per Game grouped by WinShares")
print(results)
```

## Groupby() agg() or .apply(): Group your dataset by two columns and then sort the aggregated results within the groups.

```python
new_df2 = new_df.groupby(["Years", "Plus/Minus"]).agg({"WinShares": "mean"})
```

```python
new_df3 = new_df2.apply(lambda x: x.sort_values(ascending=False))
```

```python
new_df3.head(20)
```
